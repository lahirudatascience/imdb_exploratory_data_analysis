```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the processed data
data_path = '../data/imdb_top_movies_cleaned.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the dataset
df.head()
```

## 1. Histogram for Ratings Distribution
**Chart Interpretation:**
- This histogram shows how IMDb movie ratings are distributed.
- Most movies have ratings between **8.0 and 9.0**.
- The distribution appears to be slightly skewed towards higher ratings, meaning top-rated movies dominate the dataset.
- The KDE (Kernel Density Estimation) curve helps visualize the smooth distribution of ratings.

```python
# Histogram for Ratings Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["rating"], bins=10, kde=True)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Distribution of Movie Ratings")
plt.show()
```

## 2. Histogram for Revenue Distribution
**Chart Interpretation:**
- Movie revenue follows a highly **skewed distribution**, with a few movies earning extremely high revenue.
- The majority of movies fall in a **low-revenue range**, indicating that blockbuster hits are rare.
- The use of a **logarithmic scale** on the y-axis helps better visualize variations in revenue.
- The KDE curve indicates a peak at lower revenue levels, suggesting that most movies do not earn massive amounts.

```python
# Histogram for Revenue Distribution (log scale for better visualization)
plt.figure(figsize=(10, 5))
sns.histplot(df["revenue"], bins=30, kde=True)
plt.xlabel("Revenue")
plt.ylabel("Count")
plt.title("Distribution of Movie Revenue")
plt.yscale("log")
plt.show()
```

## 3. Scatter Plot of Revenue vs Rating
**Chart Interpretation:**
- This scatter plot explores the relationship between **movie ratings and revenue**.
- There is **no strong correlation** between a movies rating and its revenue.
- Some highly-rated movies have **low revenue**, while some lower-rated movies earn significantly.
- This suggests that factors like **marketing, distribution, and audience appeal** influence revenue more than just ratings.

```python
# Scatter Plot of Revenue vs Rating
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["rating"], y=df["revenue"])
plt.xlabel("Rating")
plt.ylabel("Revenue")
plt.title("Revenue vs Rating")
plt.grid(True)
plt.show()
```

## 4. Boxplot for Revenue Over the Years
**Chart Interpretation:**
- This boxplot visualizes **how movie revenue trends over the years**.
- **Outliers (dots outside the whiskers)** indicate blockbuster movies with significantly higher earnings.
- The median revenue fluctuates, indicating that some years had more commercially successful movies.
- Older movies tend to have **lower median revenue**, but recent years have blockbuster movies pushing the upper limit.

```python
# Boxplot for Revenue over the Years
plt.figure(figsize=(12, 5))
sns.boxplot(x=df["year"], y=df["revenue"])
plt.xticks(rotation=90)
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.title("Revenue Distribution Over Years")
plt.show()
```

## 5. Boxplot for Ratings Over the Years
**Chart Interpretation:**
- This boxplot helps analyze **rating trends across different years**.
- The **median rating remains fairly stable**, meaning the quality of top movies has remained high over time.
- Some years have wider interquartile ranges (boxes), indicating **greater variation in ratings** among movies.
- Outliers suggest that some movies receive significantly higher or lower ratings than the average in that year.
- There is **no clear upward or downward trend**, indicating that highly-rated movies have consistently been produced over the decades.

```python
# Boxplot for Ratings over the Years
plt.figure(figsize=(12, 5))
sns.boxplot(x=df["year"], y=df["rating"])
plt.xticks(rotation=90)
plt.xlabel("Year")
plt.ylabel("Rating")
plt.title("Rating Distribution Over Years")
plt.show()
```

## 6. Line Plot of Average Revenue Over Time
**Chart Interpretation:**
- This line plot shows how **average movie revenue** has changed over the years.
- A general **upward trend** in revenue suggests that movies have been earning more over time, especially in recent decades.
- Some dips may indicate years where fewer blockbuster hits were released.
- The orange markers highlight yearly fluctuations.

```python
# Group by year and calculate average revenue
revenue_over_time = df.groupby('year')['revenue'].mean()

# Line plot of average revenue over time
plt.figure(figsize=(12, 6))
sns.lineplot(x=revenue_over_time.index, y=revenue_over_time.values, marker='o', color='orange')
plt.title('Average Revenue Over Time')
plt.xlabel('Year')
plt.ylabel('Average Revenue')
plt.grid(True)
plt.show()
```

## 7. Line Plot of Average Rating Over Time
**Chart Interpretation:**
- This line plot displays the **trend of average movie ratings** over time.
- The average rating appears **relatively stable** over the years, showing that high-quality movies are consistently produced.
- Some minor fluctuations suggest certain years had slightly better or worse movies on average.
- The red markers indicate year-wise ratings.

```python
# Group by year and calculate average rating
rating_over_time = df.groupby('year')['rating'].mean()

# Line plot of average rating over time
plt.figure(figsize=(12, 6))
sns.lineplot(x=rating_over_time.index, y=rating_over_time.values, marker='o', color='red')
plt.title('Average Rating Over Time')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

## 8. Bar Plot of Top 10 Highest-Grossing Movies
**Chart Interpretation:**
- This bar plot highlights the **top 10 movies with the highest revenue**.
- The **movie with the highest revenue is at the top**, and the descending order shows how revenue varies among these blockbuster hits.
- The use of the **'magma' color palette** enhances the visual appeal.
- This helps identify which movies dominated the box office historically.

```python
# Top 10 Highest-Grossing Movies
# Sort by revenue and select top 10
top_10_movies = df.sort_values(by='revenue', ascending=False).head(10)

# Bar plot of top 10 highest-grossing movies
plt.figure(figsize=(12, 6))
sns.barplot(x='revenue', y='title', data=top_10_movies, palette='magma')
plt.title('Top 10 Highest-Grossing Movies')
plt.xlabel('Revenue')
plt.ylabel('Movie Title')
plt.show()
```

## 9. Correlation Heatmap
**Chart Interpretation:**
- This heatmap shows the correlation between **revenue, rating, and release year**.
- **Revenue and year** may show some positive correlation, indicating that newer movies generally earn more.
- **Rating and revenue** do not have a strong correlation, supporting the earlier observation from the scatter plot.
- A **color gradient from blue (negative correlation) to red (positive correlation)** helps visualize these relationships.

```python
# Correlation Heatmap
# Calculate correlation matrix
corr_matrix = df[['revenue', 'rating', 'year']].corr()

# Heatmap of correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```
